```python
# Define the variable
t = var('t')

# Define the piecewise function
a = piecewise([
    ((0, 1),0),
    ((1, 2), t - 1),
    ((2, 3), -t + 3),
    ((3, 7),0),
    ((7, 8), -t + 7),
    ((8, 9), t - 9),
    ((9,10),0)
])
```

```python
plot(a(t),(t,0,10))
```

```python
# Define the piecewise function
v = piecewise([
    ((0, 1),0),
    ((1, 2), t^2/2 - t +1/2),
    ((2, 3), -t^2/2 + 3*t -7/2),
    ((3, 7),1),
    ((7, 8), -t^2/2 + 7*t -47/2),
    ((8, 9), t^2/2 - 9*t +81/2),
    ((9,10),0)
])
```

```python
plot(v,(t,0,10))
```
